Section 4: Sentiment Analysis using TextBlob

```python
import pandas as pd
from textblob import TextBlob
```

```python
#check the columns and select the column that will be apt for review
df.columns
```

```python
#the 'content' column need to be renamed to 'review'
df.rename(columns={'content': 'review'}, inplace=True)
print(df.head())
```

```python
#check once again if the column name is changed
df.columns
```

```python
def sentiment_analysis(review):
    analysis = TextBlob(review)
    if analysis.sentiment.polarity > 0:
        return 'positive'
    elif analysis.sentiment.polarity < 0:
        return 'negative'
    else:
        return 'neutral'

# Apply sentiment analysis to the 'review' column
df['sentiment'] = df['review'].apply(sentiment_analysis)

# total count and percentage of sentiment score
sentiment_counts = df['sentiment'].value_counts()
total_reviews = len(df)
sentiment_percentage = (sentiment_counts / total_reviews) * 100

# print the results
print("Sentiment Counts:\n", sentiment_counts)
print("\nSentiment Percentages:\n", sentiment_percentage)

# overall score display
df['polarity'] = df['review'].apply(lambda x: TextBlob(x).sentiment.polarity)
overall_sentiment_score = df['polarity'].mean()
print("\nOverall Sentiment Score (Polarity):", overall_sentiment_score)
```

```python
#targettig just the 'beats by dre'product to check the sentiment score

df_product = df[df['product_id'] == 'B09JL41N9C'].copy()
```

```python
def beats_sentiment(review):
    analysis = TextBlob(review)
    if analysis.sentiment.polarity > 0:
        return 'positive'
    elif analysis.sentiment.polarity == 0:
        return 'neutral'
    else:
        return 'negative'


df_product['sentiment'] = df_product['review'].apply(beats_sentiment)

#count and percentage
sentiment_counts = df_product['sentiment'].value_counts()
sentiment_percentages = df_product['sentiment'].value_counts(normalize=True) * 100

#sentiment polarity score
overall_polarity = df_product['review'].apply(lambda x: TextBlob(x).sentiment.polarity).mean()

#printing the results
print("Sentiment Counts:\n", sentiment_counts)
print("\nSentiment Percentages:\n", sentiment_percentages)
print("\nOverall Sentiment Score (Polarity):", overall_polarity)
```
